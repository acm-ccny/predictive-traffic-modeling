```python
#NYC Events Dataset
import pandas as pd
import matplotlib.pyplot as plt

eventsRead = pd.read_csv("sample_data/dataset3_cleaned.csv")

#General information about dataset
nycEvents = pd.DataFrame(eventsRead)
print(f"The colums of the NYC Events Dataset is: {nycEvents.columns} \n")
print(f"The shape of the NYC Events Dataset is: {nycEvents.shape} \n")
print(f"The statistical data on NYC Events Dataset is: {nycEvents.describe()} \n")
print(f"The info on Nyc Events Dataset is: {nycEvents.info()} \n")

#Event Street Side" and "Street Closure Type" columns have the highest Nan data types
nycColumnList = [nycEvents.columns]
for Column in nycColumnList:
   print(f"The total missing data in all the columns is: {nycEvents[Column].isna().sum()}")


nycEvents = nycEvents.drop_duplicates()
nycEvents
```

```python
eventsRead2 = pd.read_csv("sample_data/dataset4_cleaned.csv")

#General information about dataset
nycEventsHist = pd.DataFrame(eventsRead2)
print(f"The colums of the NYC Events Dataset is: {nycEventsHist.columns} \n")
print(f"The shape of the NYC Events Dataset is: {nycEventsHist.shape} \n")
print(f"The statistical data on NYC Events Dataset is: {nycEventsHist.describe()} \n")
print(f"The info on Nyc Events Dataset is: {nycEventsHist.info()} \n")


nycColumnList2 = [nycEventsHist.columns]
for Column in nycColumnList2:
   print(f"The total missing data in all the columns is: {nycEventsHist[Column].isna().sum()}")


nycEventsHist = nycEventsHist.drop_duplicates()
nycEventsHist
```

```python
#Merging
from google.colab import files
import pandas as pd


# combine on Event ID
# Combine creates more columns in dataset through stacking data.
# The variables nycEvents and nycEventsHist are currently method objects because .drop_duplicates was assigned, not called.
# We need to call these methods to get the DataFrames and use pd.concat for combining.
combined = pd.concat([nycEvents(), nycEventsHist()], ignore_index=True)
```

```python
#info on combined datasef

combined = combined.drop_duplicates()



#General information about dataset
print(f"The colums of the combined dataset is: {combined.columns} \n")
print(f"The shape of the combined dataset is: {combined.shape} \n")
print(f"The statistical data on combined dataset is: {combined.describe()} \n")
print(f"The info on the combined dataset is: {combined.info()} \n")


combinedColumnList = [combined.columns]
for column in combinedColumnList:
   print(f"The total missing data in the combined dataset is:\n{combined[column].isna().sum()}")


combined
```

```python
# convert dates
combined["Start Date/Time"] = pd.to_datetime(combined["Start Date/Time"])
combined["End Date/Time"] = pd.to_datetime(combined["End Date/Time"])

# remove strong duplicates
combined = combined.drop_duplicates(
    subset=["Event Name", "Start Date/Time", "End Date/Time"]
)

# check missing
print(combined.isna().sum().sort_values(ascending=False))

from google.colab import files

# save updated dataset
combined.to_csv("cleaned_combined_dataset.csv", index=False)

# download it
files.download("cleaned_combined_dataset.csv")
```
