```python
import pandas as pd

dataset1 = pd.read_csv("sample_data/dataset1_cleaned.csv")
dataset2 = pd.read_csv("sample_data/dataset2_cleaned.csv")
dataset6 = pd.read_csv("sample_data/dataset6_cleaned.csv")

borough_map = {
    "Manhattan": "Manhattan",
    "New York": "Manhattan",
    "Bronx": "Bronx",
    "The Bronx": "Bronx",
    "Bronx County": "Bronx",
    "Brooklyn": "Brooklyn",
    "Queens": "Queens",
    "Staten Island": "Staten Island",
    "Staten island": "Staten Island"
}

for df in [dataset1, dataset2, dataset6]:
    df["borough"] = df["borough"].astype(str).str.strip().str.title()
    df["borough"] = df["borough"].map(borough_map).fillna(df["borough"])

dataset1 = dataset1.dropna(subset=["borough"])
dataset2 = dataset2.dropna(subset=["borough"])
dataset6 = dataset6.dropna(subset=["borough"])

dataset6["datetime"] = pd.to_datetime(
    dict(
        year=dataset6["Yr"],
        month=dataset6["M"],
        day=dataset6["D"],
        hour=dataset6["HH"],
        minute=dataset6["MM"]
    ),
    errors="coerce"
)

dataset6 = dataset6.dropna(subset=["datetime"])

dataset1_borough = dataset1.groupby("borough", as_index=False).mean(numeric_only=True)
dataset2_borough = dataset2.groupby("borough", as_index=False).mean(numeric_only=True)

merged_dataset = pd.merge(dataset6, dataset1_borough, on="borough", how="left")
merged_dataset = pd.merge(merged_dataset, dataset2_borough, on="borough", how="left")

merged_dataset = merged_dataset.sort_values(
    by=["borough", "street", "fromSt", "toSt", "datetime"]
)

cols = ["borough", "street", "fromSt", "toSt", "datetime"] + [
    col for col in merged_dataset.columns
    if col not in ["borough", "street", "fromSt", "toSt", "datetime"]
]

merged_dataset = merged_dataset[cols]

print("dataset1 boroughs:", sorted(dataset1["borough"].dropna().unique()))
print("dataset2 boroughs:", sorted(dataset2["borough"].dropna().unique()))
print("dataset6 boroughs:", sorted(dataset6["borough"].dropna().unique()))
print("merged boroughs:", sorted(merged_dataset["borough"].dropna().unique()))

print(merged_dataset.head(20))
print(merged_dataset.shape)
print(merged_dataset.isnull().sum())

merged_dataset.to_csv("merged_dataset.csv", index=False)
```
