#### Loading Raw Dataset - [NYC Permitted Event Information](https://data.cityofnewyork.us/City-Government/NYC-Permitted-Event-Information/tvpp-9vvx/about_data)



```python
# Loads the dataset into a pandas DataFrame
dataset5raw = pd.read_csv('../../data/raw/Street_Closures_due_to_Construction_Activities_by_Block_and_Intersection_20260321.csv')  
# Displays the first few rows to verify import happened
dataset5raw.head()  
```

#### Inspect Dataset for if it needs to be cleaned or not

```python
# Look at the general information of the datset to understand its setup and structure
dataset5raw.info()
```

```python
# Total datapoints in the dataset
total_datapoints = len(dataset5raw)

# Checks for the amount of missing values in each column
missingValsCount = dataset5raw.isnull().sum()
print("Missing values data percentage:\n" + str(missingValsCount / total_datapoints * 100), end="")
print("\n\nMissing values count:\n" + str(missingValsCount), end="")

# Checks for the amount of duplicated rows in the dataset
duplicatedValsCount = (dataset5raw.duplicated().sum())
print("\n\nDuplicated rows data percentage:", duplicatedValsCount / total_datapoints * 100)
```

Conclusion Reached:

The main issue with this dataset is that ToStreetName is missing many values. The assumption can be made that this event menas only that one street name is being taken. 


#### Dataset Cleaning

```python
# Adjust the columns involving dates and times to be datetime objects 
dataset5raw['WorkStartDate'] = pd.to_datetime(dataset5raw['WorkStartDate'])
dataset5raw['WorkEndDate'] = pd.to_datetime(dataset5raw['WorkEndDate'])

# Cleans Event Street Side and Street Closure Type to have empty strings instead of null values and converts them to string type
dataset5raw['FromStreetName'] = dataset5raw['FromStreetName'].fillna('').astype(str)
dataset5raw['ToStreetName'] = dataset5raw['ToStreetName'].fillna('').astype(str)

dataset5raw.info()
```

Exports cleaned data to data/processed 

```python
dataset5raw.to_csv("../../data/processed/dataset5_cleaned.csv", index=False)
```
