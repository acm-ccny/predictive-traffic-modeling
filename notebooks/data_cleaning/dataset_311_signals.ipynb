# Clean Dataset
In this section, we clean each dataset by removing duplicates, handling missing or invalid values, standardizing column names and formats, correcting data types, and ensuring that columns intended for merging are consistent across datasets.

```python
# remove duplicate rows
# example:
# dataset1.drop_duplicates(inplace=True)

# WRITE YOUR CODE HERE
dataset_311_signals.drop_duplicates(inplace=True)
```

```python
# handle missing or invalid values
# options include:
# - dropping rows: dataset1.dropna(inplace=True)
# - filling values: dataset1['column_name'].fillna(value, inplace=True)

# WRITE YOUR CODE HERE

# drop columns with extremely high missing values
dataset_311_signals.drop(columns=["Bridge Highway Name", "Bridge Highway Direction"], inplace=True)

dataset_311_signals["City"].fillna("Unknown", inplace=True)
```

```python
# standardize column names and formats
# example:
# dataset1.columns = dataset1.columns.str.lower().str.replace(" ", "_")

# WRITE YOUR CODE HERE
dataset_311_signals.columns = dataset_311_signals.columns.str.lower().str.replace(" ", "_") 
```

```python
# correct data types as needed
# example:
# dataset1['date'] = pd.to_datetime(dataset1['date'])
# dataset1['value'] = dataset1['value'].astype(float)

# WRITE YOUR CODE HERE
dataset_311_signals["created_date"] = pd.to_datetime(dataset_311_signals["created_date"])
dataset_311_signals["closed_date"] = pd.to_datetime(dataset_311_signals["closed_date"])
```
