```python
import pandas as pd

readData1 = pd.read_csv("sample_data/Dataset(1).csv")
readData2 = pd.read_csv("sample_data/Dataset(2).csv")
readData6 = pd.read_csv("sample_data/dataset6_cleaned.csv", )

dataset1 = pd.DataFrame(readData1)
dataset2 = pd.DataFrame(readData2)
dataset6 = pd.DataFrame(readData6)


############Checking information on dataframes.#################

print(f"Columns in dataset1: {dataset1.columns}")
print(f"Shape of dataset1: {dataset1.shape}\n")
print(f"dataset1 missing vals:\n{dataset1.isnull().sum()}/n")
print(f"dataset1 statistical summarization:\n{dataset1.describe()}\n")
print(f"dataset1 information:\n{dataset1.info()}")
print(f"dataset1 head\n:{dataset1.head()}\n\n\n")

print(f"Colums in dataset2: {dataset2.columns}")
print(f"Shape of dataset2: {dataset2.shape}\n")
print("dataset2 missing vals:\n", dataset2.isnull().sum(),"\n")
print(f"dataset2 statistical summarization:\n{dataset2.describe()}\n")
print(f"dataset2 information:\n{dataset2.info()}")
print(f"dataset2 head\n:{dataset2.head()}\n\n\n")


print(f"Colums in dataset6: {dataset6.columns}")
print(f"Shape of dataset6: {dataset6.shape}\n")
print("dataset6 missing vals:\n", dataset6.isnull().sum(),"\n")
print(f"dataset6 statistical summarization:\n{dataset6.describe()}\n")
print(f"dataset6 information:\n{dataset6.info()}")
print(f"dataset6 head\n:{dataset6.head()}\n\n\n")
#finding dataset6 rows with missing values
print(dataset6[dataset6.isnull().any(axis=1)])





print("End of general information section\n")

###########Cleaning datasets################################

#Cleaning dataset 1

print("Cleanding datasets section\n")
dataset1 = dataset1.drop_duplicates()


#Cleaning dataset 2
dataset2 = dataset2.drop_duplicates()
# shorten it (overwrite original)
dataset2 = dataset2.rename(columns={"Boro": "borough"})
dataset2 = dataset2.groupby("borough", group_keys=False).apply(
    lambda x: x.sample(min(len(x), 4000), random_state=42)
).reset_index(drop=True)




#Cleaning dataset 6

#dropping rows with missing vals from datset6 since only 1276 have Nan out of 1875154)
dataset6 = dataset6.dropna(subset=["toSt"])

dataset6 = dataset6.drop_duplicates()
dataset6 = dataset6.dropna(subset=["Direction"])
dataset6["Vol"] = pd.to_numeric(dataset6["Vol"], errors="coerce")
dataset6 = dataset6.dropna(subset=["Vol"])
dataset6 = dataset6.rename(columns = {"Boro": "borough"})
print(dataset6.head())
print(dataset6.isnull().sum())





```

```python
dataset1.to_csv("dataset1_cleaned.csv", index=False)
dataset2.to_csv("dataset2_cleaned.csv", index=False)
dataset6.to_csv("dataset6_cleaned.csv", index=False)

from google.colab import files

files.download("dataset1_cleaned.csv")
files.download("dataset2_cleaned.csv")
files.download("dataset6_cleaned.csv")
```
