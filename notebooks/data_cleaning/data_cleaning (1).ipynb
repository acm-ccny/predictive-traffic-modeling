```python
#Loading Manhattan EZ pass data set

import pandas as pd

#import polyline

#NOTE: \n is to create a new line in output.

#To read csv file of EZ pass data set
manhattanEZPass = pd.read_csv("NYC OpenData EZ Pass Reader (Midtown Manhattan).csv")

########### Data on original dataset ###########

print("Info on dataframe.", manhattanEZPass.info(), '\n')

print("Stats on dataframe.", manhattanEZPass.describe(), '\n')

print("Number of rows and columns: ",manhattanEZPass.shape, '\n') #rows and columns of data set

#Names of all the columns of the dataframe
print("Names of columns: ", manhattanEZPass.columns, '\n')

print("Number of missing values:", '\n', manhattanEZPass.isnull().sum(), '\n')

#median_calculation_timestamp shows the date of the received data (useful to know for merging)
print("Timestamps of data column:", manhattanEZPass["median_calculation_timestamp"], "\n")

print("EZ Pass DataFrame:")
print(manhattanEZPass.head(), '\n', '\n')





############ Cleaning Dataset.  #################

#Remove duplicate rows in the dataset
manhattanEZPass.drop_duplicates(inplace = True)

#Standardizing column names (making column names all lowercase)
manhattanEZPass.columns = manhattanEZPass.columns.str.lower()

print("Cleaned EZ Pass Dtaset: ")


manhattanEZPass.head()











```
